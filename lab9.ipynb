```python
import csv

class ExpenseTracker:
    def __init__(self):
        self.transactions = []

    def add_transaction(self, transaction_type, amount):
        self.transactions.append({"Type": transaction_type, "Amount": amount})

    def total_expense(self):
        return sum(item["Amount"] for item in self.transactions if item["Type"] == "Expense")

    def total_income(self):
        return sum(item["Amount"] for item in self.transactions if item["Type"] == "Income")

    def save_to_csv(self, filename):
        with open(filename, mode='w', newline='') as file:
            fieldnames = ['Type', 'Amount']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            for transaction in self.transactions:
                writer.writerow(transaction)

    def load_from_csv(self, filename):
        self.transactions = []
        with open(filename, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                self.transactions.append(row)

def main():
    tracker = ExpenseTracker()
    
    # Load data from a CSV file
    csv_file = "expenses.csv"
    tracker.load_from_csv(csv_file)

    while True:
        print("\nExpense Tracker Menu:")
        print("1. Add Expense")
        print("2. Add Income")
        print("3. Show Total Expenses")
        print("4. Show Total Income")
        print("5. Save and Exit")
        
        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            amount = float(input("Enter the expense amount: "))
            tracker.add_transaction("Expense", amount)
            print(f"Expense of ${amount} added.")

        elif choice == '2':
            amount = float(input("Enter the income amount: "))
            tracker.add_transaction("Income", amount)
            print(f"Income of ${amount} added.")

        elif choice == '3':
            total_expense = tracker.total_expense()
            print(f"Total Expenses: ${total_expense}")

        elif choice == '4':
            total_income = tracker.total_income()
            print(f"Total Income: ${total_income}")

        elif choice == '5':
            # Save data to CSV and exit
            tracker.save_to_csv(csv_file)
            print("Data saved to 'expenses.csv'. Exiting.")
            break

        else:
            print("Invalid choice. Please select a valid option (1/2/3/4/5).")

if __name__ == "__main__":
    main()
```

Matrix Multiplication

Jeeva, a mathematical expert, is looking for an expert in programming who can assist him in creating a Python program for Matrix Multiplication.

His program has specific requirements.

The matrices have to be saved inside a dictionary.
The keys allowed in the dictionary are [matrix 1, matrix 2, result matrix]
The values for the elements in the rows of the matrix have to be saved in the form of a list of lists [[row1][row2]].
You must have functions for collecting the matrix and store in the dictionary.
Another function to Process the matrix and store it in the dictionary. 
Another function to Display the result in a file, properly formatted.
Menu option to show enter matrix 1, matrix 2, calculate, export to file. If the matrix has been entered, then it should take confirmation from the user before taking the input for matrix 1 or 2 again.
Create lab 08.py file or .ipynb file, and upload in GitHub.

```python
import numpy as np

# Initialize the dictionary to store matrices
matrix_dict = {
    "matrix 1": [],
    "matrix 2": [],
    "result matrix": []
}

# Function to collect and store a matrix in the dictionary
def collect_matrix(matrix_name):
    matrix = []
    while True:
        try:
            rows = int(input(f"Enter the number of rows for {matrix_name}: "))
            columns = int(input(f"Enter the number of columns for {matrix_name}: "))
            break
        except ValueError:
            print("Please enter a valid number of rows and columns.")

    for i in range(rows):
        row = []
        for j in range(columns):
            while True:
                try:
                    element = float(input(f"Enter element at position ({i+1}, {j+1}): "))
                    break
                except ValueError:
                    print("Please enter a valid number.")
            row.append(element)
        matrix.append(row)
    
    matrix_dict[matrix_name] = matrix

# Function to perform matrix multiplication and store the result in the dictionary
def perform_matrix_multiplication():
    if not matrix_dict["matrix 1"] or not matrix_dict["matrix 2"]:
        print("Please enter both matrices before performing multiplication.")
        return
    
    matrix1 = np.array(matrix_dict["matrix 1"])
    matrix2 = np.array(matrix_dict["matrix 2"])
    
    try:
        result_matrix = np.dot(matrix1, matrix2)
        matrix_dict["result matrix"] = result_matrix.tolist()
    except ValueError as e:
        print("Matrix multiplication is not possible with the given matrices:", e)

# Function to display the result in a file
def export_result_to_file(filename):
    result_matrix = matrix_dict["result matrix"]
    if not result_matrix:
        print("No result to export. Perform matrix multiplication first.")
        return
    
    with open(filename, "w") as file:
        for row in result_matrix:
            file.write("\t".join(map(str, row)) + "\n")

# Main menu
while True:
    print("\nMatrix Multiplication Program")
    print("1. Enter Matrix 1")
    print("2. Enter Matrix 2")
    print("3. Calculate Matrix Multiplication")
    print("4. Export Result to File")
    print("5. Quit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        collect_matrix("matrix 1")
    elif choice == "2":
        collect_matrix("matrix 2")
    elif choice == "3":
        perform_matrix_multiplication()
    elif choice == "4":
        filename = input("Enter the filename to export the result (e.g., result.txt): ")
        export_result_to_file(filename)
    elif choice == "5":
        break
    else:
        print("Invalid choice. Please select a valid option.")
```

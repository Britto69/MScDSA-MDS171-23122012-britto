As Culinary Delights gained popularity, it faced a significant challenge â€“ managing orders efficiently. Chef Amelia realised that she needed a robust Restaurant Management System (RMS) to maintain the restaurant's reputation.

```python
menu = {
    "Fried Rice": 200,
    "Butter Chicken": 250,
    "Rice": 50,
    "Tea": 10
}

# Function to display the menu
def display_menu():
    print("\nMenu:")
    for item, price in menu.items():
        print(f"{item}: Rs. {price}")

# Function to record a new order
def record_order(customer_name, order_items, total_amount):
    with open("restaurant_orders.txt", "a") as file:
        file.write(f"Customer Name: {customer_name}\n")
        file.write("Order Items:\n")
        for item, quantity in order_items.items():
            file.write(f"- {item}: {quantity} x Rs. {menu[item]} = Rs. {quantity * menu[item]}\n")
        file.write(f"Total Amount: Rs. {total_amount}\n")
        file.write("=" * 40 + "\n")

# Function to record customer review
def record_review(customer_name, review):
    with open("restaurant_reviews.txt", "a") as file:
        file.write(f"Customer Name: {customer_name}\n")
        file.write(f"Review: {review}\n")
        file.write("=" * 40 + "\n")

if __name__ == "__main__":
    while True:
        print("\nRestaurant Management System")
        print("1. Display Menu")
        print("2. Place an Order")
        print("3. Leave a Review")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            display_menu()

        elif choice == "2":
            customer_name = input("Customer Name: ")
            order_items = {}
            total_amount = 0

            while True:
                display_menu()
                item = input("Enter item from the menu (or 'done' to finish): ")
                if item == 'done':
                    break
                if item in menu:
                    quantity = int(input(f"Enter quantity of {item}: "))
                    order_items[item] = quantity
                    total_amount += quantity * menu[item]
                else:
                    print("Item not found in the menu.")

            record_order(customer_name, order_items, total_amount)
            print("Order recorded successfully!")

        elif choice == "3":
            customer_name = input("Customer Name: ")
            review = input("Enter your review: ")

            record_review(customer_name, review)
            print("Review recorded successfully!")

        elif choice == "4":
            break
  
```
